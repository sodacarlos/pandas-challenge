```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data_df = pd.read_csv(school_data_to_load)
student_data_df = pd.read_csv(student_data_to_load, dtype = {'year': str})

# Combine the data into a single dataset.  
school_data_complete_df = pd.merge(student_data_df, school_data_df, how="left", on=["school_name", "school_name"])
school_data_complete_df.head()
```

## Local Government Area Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average maths score 

* Calculate the average reading score

* Calculate the percentage of students with a passing maths score (50 or greater)

* Calculate the percentage of students with a passing reading score (50 or greater)

* Calculate the percentage of students who passed maths **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Calculate the Totals (Schools and Students)
school_count = len(school_data_complete_df["school_name"].unique())
student_count = school_data_complete_df["Student ID"].count()

# Calculate the Total Budget
total_budget = school_data_df["budget"].sum()

# Calculate the Average Scores.
average_reading_score = school_data_complete_df["reading_score"].mean()
average_maths_score = school_data_complete_df["maths_score"].mean()

# Calculate the passing rates.
passing_maths_count = school_data_complete_df[(school_data_complete_df["maths_score"] >= 50)].count()["student_name"]
passing_reading_count = school_data_complete_df[(school_data_complete_df["reading_score"] >= 50)].count()["student_name"]

# Calculate the passing percentages.
passing_reading_percentage = passing_reading_count/student_count*100
passing_maths_percentage = passing_maths_count/student_count*100

# Calculate the students who passed both reading and maths.
passing_maths_reading = school_data_complete_df[(school_data_complete_df["maths_score"] >= 50)
                                               & (school_data_complete_df["reading_score"] >= 50)]
# Calculate the number of students that passed both reading and maths.
overall_passing_maths_reading_count = passing_maths_reading["student_name"].count()

# Calculate the overall passing percentage with total student count.
overall_passing_percentage = overall_passing_maths_reading_count / student_count*100

# Create a DataFrame
LGA_summary_df = pd.DataFrame(
          [{"Total Schools": school_count, 
          "Total Students": student_count, 
          "Total Budget": total_budget,
          "Average Maths Score": average_maths_score, 
          "Average Reading Score": average_reading_score,
          "% Passing Maths": passing_maths_percentage,
         "% Passing Reading": passing_reading_percentage,
        "% Overall Passing": overall_passing_percentage}])

# Formats:
LGA_summary_df["Total Students"] = LGA_summary_df["Total Students"].map("{:,}".format)
LGA_summary_df["Total Budget"] = LGA_summary_df["Total Budget"].map("${:,.2f}".format)
LGA_summary_df["Average Maths Score"] = LGA_summary_df["Average Maths Score"].map("{:.2f}".format)
LGA_summary_df["Average Reading Score"] = LGA_summary_df["Average Reading Score"].map("{:.2f}".format)
LGA_summary_df["% Passing Maths"] = LGA_summary_df["% Passing Maths"].map("{:.2f}".format)
LGA_summary_df["% Passing Reading"] = LGA_summary_df["% Passing Reading"].map("{:.2f}".format)
LGA_summary_df["% Overall Passing"] = LGA_summary_df["% Overall Passing"].map("{:.2f}".format)

# Display the data frame
LGA_summary_df
```

## School Summary

* Create an overview table that summarises key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed maths **and** reading.)
  
* Create a dataframe to hold the above results

```python
# School Summary

# Determine the School Type
Sschool_types = school_data_df.set_index(["school_name"])["type"]
# Calculate the total student count.
Sschool_counts = school_data_complete_df["school_name"].value_counts()
# Calculate the total school budget and per capita spending
Sschool_budget = school_data_complete_df.groupby(["school_name"]).mean()["budget"]
# Calculate the per capita spending.
Sschool_capita = Sschool_budget / Sschool_counts
# Calculate the average test scores.
Sschool_maths = school_data_complete_df.groupby(["school_name"]).mean()["maths_score"]
Sschool_reading = school_data_complete_df.groupby(["school_name"]).mean()["reading_score"]
# Calculate the passing scores by creating a filtered DataFrame.
Sschool_passing_maths = school_data_complete_df[(school_data_complete_df["maths_score"] >= 50)]
Sschool_passing_reading = school_data_complete_df[(school_data_complete_df["reading_score"] >= 50)]
# Calculate the number of students passing math and passing reading by school.
Sschool_passing_maths = Sschool_passing_maths.groupby(["school_name"]).count()["student_name"]
Sschool_passing_reading = Sschool_passing_reading.groupby(["school_name"]).count()["student_name"]
# Calculate the percentage of passing maths and reading scores per school.
Sschool_passing_maths = Sschool_passing_maths / Sschool_counts * 100
Sschool_passing_reading = Sschool_passing_reading / Sschool_counts * 100
# Calculate the students who passed both reading and maths.
Spassing_maths_reading = school_data_complete_df[(school_data_complete_df["reading_score"] >= 50)
    & (school_data_complete_df["maths_score"] >= 50)]

# Calculate the number of students passing math and passing reading by school.
Spassing_maths_reading = Spassing_maths_reading.groupby(["school_name"]).count()["student_name"]
# Calculate the percentage of passing math and reading scores per school.
Soverall_passing_percentage = Spassing_maths_reading/Sschool_counts * 100

# Create the DataFrame
school_summary_df = pd.DataFrame({
    "School Type": Sschool_types,
    "Total Students": Sschool_counts,
    "Total School Budget": Sschool_budget,
    "Per Student Budget": Sschool_capita,
    "Average Maths Score": Sschool_maths,
    "Average Reading Score": Sschool_reading,
    "% Passing Maths": Sschool_passing_maths,
    "% Passing Reading": Sschool_passing_reading,
    "% Overall Passing": Soverall_passing_percentage})

# Formats:
school_summary_df["Total School Budget"] = school_summary_df["Total School Budget"].map("${:,.2f}".format)
school_summary_df["Per Student Budget"] = school_summary_df["Per Student Budget"].map("${:,.2f}".format)

# Display the data frame
school_summary_df
```

## Maths Scores by Year

* Create a table that lists the average maths score for students of each year level (9, 10, 11, 12) at each school.

  * Create a pandas series for each year. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
# Math and Reading Scores by Grade

# Create a Series of scores by year levels using conditionals.
year_nine = school_data_complete_df[(school_data_complete_df["year"] == "9")]
year_ten = school_data_complete_df[(school_data_complete_df["year"] == "10")]
year_eleven = school_data_complete_df[(school_data_complete_df["year"] == "11")]
year_twelve = school_data_complete_df[(school_data_complete_df["year"] == "12")]

# Group each school Series by the school name for the average maths score.
year_nine_maths_scores = year_nine.groupby(["school_name"]).mean()["maths_score"]
year_ten_maths_scores = year_ten.groupby(["school_name"]).mean()["maths_score"]
year_eleven_maths_scores = year_eleven.groupby(["school_name"]).mean()["maths_score"]
year_twelve_maths_scores = year_twelve.groupby(["school_name"]).mean()["maths_score"]

# Group each school Series by the school name for the average reading score.
year_nine_reading_scores = year_nine.groupby(["school_name"]).mean()["reading_score"]
year_ten_reading_scores = year_ten.groupby(["school_name"]).mean()["reading_score"]
year_eleven_reading_scores = year_eleven.groupby(["school_name"]).mean()["reading_score"]
year_twelve_reading_scores = year_twelve.groupby(["school_name"]).mean()["reading_score"]

# Combine each Series for average math scores by school into single data frame.
maths_scores_by_year = pd.DataFrame({
                "Year 9": year_nine_maths_scores,
               "Year 10": year_ten_maths_scores,
               "Year 11": year_eleven_maths_scores,
               "Year 12": year_twelve_maths_scores})

# Combine each Series for average reading scores by school into single data frame.
reading_scores_by_year = pd.DataFrame({
               "Year 9": year_nine_reading_scores,
              "Year 10": year_ten_reading_scores,
              "Year 11": year_eleven_reading_scores,
              "Year 12": year_twelve_reading_scores})

# Formats:
maths_scores_by_year["Year 9"] = maths_scores_by_year["Year 9"].map("{:.1f}".format)
maths_scores_by_year["Year 10"] = maths_scores_by_year["Year 10"].map("{:.1f}".format)
maths_scores_by_year["Year 11"] = maths_scores_by_year["Year 11"].map("{:.1f}".format)
maths_scores_by_year["Year 12"] = maths_scores_by_year["Year 12"].map("{:.1f}".format)
maths_scores_by_year = maths_scores_by_year[["Year 9","Year 10", "Year 11", "Year 12"]]
reading_scores_by_year["Year 9"] = reading_scores_by_year["Year 9"].map("{:.1f}".format)
reading_scores_by_year["Year 10"] = reading_scores_by_year["Year 10"].map("{:.1f}".format)
reading_scores_by_year["Year 11"] = reading_scores_by_year["Year 11"].map("{:.1f}".format)
reading_scores_by_year["Year 12"] = reading_scores_by_year["Year 12"].map("{:.1f}".format)
reading_scores_by_year = reading_scores_by_year[["Year 9","Year 10", "Year 11", "Year 12"]]
maths_scores_by_year.index.name = None
reading_scores_by_year.index.name = None

# Display the data frame
maths_scores_by_year
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Scores by School Spending

# Establish the spending bins and group names.
spending_bins = [0, 585, 630, 645, 680]
group_names = ["<$584", "$585-629", "$630-644", "$645-680"]

# Categorize spending based on the bins.
school_summary_df["Spending Ranges (Per Student)"] = pd.cut(Sschool_capita, spending_bins, labels=group_names)
school_summary_df

# Calculate averages for the desired columns. 
spending_maths_scores = school_summary_df.groupby(["Spending Ranges (Per Student)"]).mean()["Average Maths Score"]
spending_reading_scores = school_summary_df.groupby(["Spending Ranges (Per Student)"]).mean()["Average Reading Score"]
spending_passing_maths = school_summary_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Maths"]
spending_passing_reading = school_summary_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Reading"]
overall_passing_spending = school_summary_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Overall Passing"]

# Create the DataFrame
spending_summary_df = pd.DataFrame({
          "Average Maths Score" : spending_maths_scores,
          "Average Reading Score": spending_reading_scores,
          "% Passing Maths": spending_passing_maths,
          "% Passing Reading": spending_passing_reading,
          "% Overall Passing": overall_passing_spending})

# Formats:
spending_summary_df["Average Maths Score"] = spending_summary_df["Average Maths Score"].map("{:.1f}".format)
spending_summary_df["Average Reading Score"] = spending_summary_df["Average Reading Score"].map("{:.1f}".format)
spending_summary_df["% Passing Maths"] = spending_summary_df["% Passing Maths"].map("{:.2f}".format)
spending_summary_df["% Passing Reading"] = spending_summary_df["% Passing Reading"].map("{:.2f}".format)
spending_summary_df["% Overall Passing"] = spending_summary_df["% Overall Passing"].map("{:.2f}".format)
spending_summary_df
```
